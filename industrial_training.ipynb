# <h1><center>Restaurant Reviews</center></h1>

# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#  Importing the dataset

```python
dataset = pd.read_csv('Restaurant_Reviews.tsv', delimiter = '\t', quoting = 3)
# with quoting=3, we are ignoring double quotes so they don't cause any problem later
```

```python
# Press shift + tab on function name if want to see its info
```

```python
dataset.head()
```

# Cleaning the texts

```python
# corpus in nlp is a collection of  text that can be anything
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 1000):
    review = re.sub('[^a-zA-Z]',' ', dataset['Review'][i])  # Numbers and exclamation marks are removed
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    # with set function it will be way a faster algorithm 
    review = ' '.join(review)
    corpus.append(review)
 
```

```python
   # After cleaning data looks like
corpus[:10]
```

# Creating the Bag of Words model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
# We will just keep 1500 most frequent words in corpus here.
X = cv.fit_transform(corpus).toarray()   
# We will make sparse matrix and then convert to array
y = dataset.iloc[:, 1].values
y
# Here X is the sparse matrix and y is dependent vector
```

#  Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
help(train_test_split)
```

#  Fitting Naive Bayes to the Training set

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
help(classifier.fit)
```

#  Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

#  Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
Accuracy=(55+91)/200
print(Accuracy)
```

```python
# Thus we get 73 % Accuracy
```
